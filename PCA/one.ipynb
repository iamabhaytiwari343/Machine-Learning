```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset for demonstration
iris = load_iris()
X = iris.data  # Features
y = iris.target  # Target variable

# Applying PCA with 2 components (you can choose the number of components)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Visualizing the results
plt.figure(figsize=(8, 6))
colors = ['navy', 'turquoise', 'darkorange']
lw = 2

for color, i, target_name in zip(colors, [0, 1, 2], iris.target_names):
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1], color=color, alpha=.8, lw=lw,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of IRIS dataset')

plt.show()
```

```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset for demonstration
iris = load_iris()
X = iris.data  # Features

# Applying PCA
pca = PCA()
X_pca = pca.fit_transform(X)

# Explained variance ratio
explained_variance = pca.explained_variance_ratio_
plt.bar(range(len(explained_variance)), explained_variance)
plt.xlabel('Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance Ratio per Principal Component')
plt.show()
```

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

# Load the Iris dataset for demonstration
iris = load_iris()
X = iris.data  # Features

# Applying PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Reconstruction of data from reduced dimensions
X_reconstructed = pca.inverse_transform(X_pca)

# Compare original data with reconstructed data
print("Original Data:")
print(X[:5])  # Displaying first 5 rows of original data

print("\nReconstructed Data:")
print(X_reconstructed[:5])  # Displaying first 5 rows of reconstructed data
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

# Load the Iris dataset for demonstration
iris = load_iris()
X = iris.data  # Features

# Applying PCA for different number of components
pca_var = []
for i in range(1, 5):  # Trying with 1 to 4 components
    pca = PCA(n_components=i)
    pca.fit(X)
    pca_var.append(np.sum(pca.explained_variance_ratio_))

# Plotting the cumulative explained variance
plt.plot(range(1, 5), pca_var, marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance vs. Number of Components')
plt.show()
```

```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Generate some example data
np.random.seed(42)
data = np.random.rand(100, 2)  # 100 samples with 2 features each

# Apply PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(data)

# Access explained variance ratio to understand the importance of each principal component
explained_variance_ratio = pca.explained_variance_ratio_
print("Explained Variance Ratio:", explained_variance_ratio)

# Plot the original data and the principal components
plt.scatter(data[:, 0], data[:, 1], label='Original Data')
plt.quiver(0, 0, pca.components_[0, 0], pca.components_[0, 1], scale=3, color='r', label='Principal Component 1')
plt.quiver(0, 0, pca.components_[1, 0], pca.components_[1, 1], scale=3, color='g', label='Principal Component 2')
plt.title('PCA Example')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
